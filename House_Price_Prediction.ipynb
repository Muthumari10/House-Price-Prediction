```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('train.csv')

data.head()

print(data.columns)

data= data[['BedroomAbvGr','GrLivArea','BsmtFullBath', 'BsmtHalfBath', 'FullBath',
       'HalfBath','SalePrice']]

data= data.fillna(0)

data['TotalBathrooms']= ( data['BsmtFullBath']+data['FullBath']+ 0.5*data['BsmtHalfBath'] + 0.5*data['HalfBath'])

X= data[['TotalBathrooms','BedroomAbvGr','GrLivArea']]

Y= data['SalePrice']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

print(f"Mean Squared Error:" ,mean_squared_error(Y_test, Y_pred))
print(f"R2 Score:" ,r2_score(Y_test, Y_pred))

features_names = ['TotalBathrooms','BedroomAbvGr','GrLivArea']
totalbathrooms = float(input("Enter Total No of Bathrooms"))
bedrooms = float(input("Enter Total No of Bedrooms"))
area = float(input("Enter the living area (in sq ft)"))
house = pd.DataFrame([[totalbathrooms,bedrooms,area]], columns = features_names)
predict_price = model.predict(house)
print(predict_price[0])

plt.scatter(Y_test, Y_pred)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.show()

import pickle
with open('house_model.pkl', 'wb') as file:
    pickle.dump(model, file)  # This saves the model


```
